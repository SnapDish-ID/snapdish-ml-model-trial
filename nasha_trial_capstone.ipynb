```python
import tensorflow as tf
from tensorflow.keras import layers

batch_size = 32
img_size = (150, 150)

base_dir = "dataset/Vegetable Images"

# Load the training dataset
train_dataset = tf.keras.utils.image_dataset_from_directory(
    directory=f"{base_dir}/train",
    image_size=img_size,
    batch_size=batch_size,
)

# Load the validation dataset
val_dataset = tf.keras.utils.image_dataset_from_directory(
    directory=f"{base_dir}/validation",
    image_size=img_size,
    batch_size=batch_size,
)

# Load the test dataset
test_dataset = tf.keras.utils.image_dataset_from_directory(
    directory=f"{base_dir}/test",
    image_size=img_size,
    batch_size=batch_size,
)

# Verification
print(f"Train dataset: {train_dataset}")
print(f"Validation dataset: {val_dataset}")
print(f"Test dataset: {test_dataset}")
```

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout

base_model = MobileNetV2(input_shape=(150, 150, 3), include_top=False, weights='imagenet')
base_model.trainable = False  # Freeze layer pre-trained

model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(len(train_dataset.class_names), activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
epochs = 10
history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=epochs,
    # callbacks = [EarlyStoppingCallback()]
)
```

```python
import os
from PIL import Image

# Direktori dataset yang sudah diekstrak
dataset_dir = 'dataset'
output_dir = 'resized_dataset'

# Membuat folder output utama jika belum ada
os.makedirs(output_dir, exist_ok=True)

# Ukuran baru untuk resize
new_size = (224, 224)

# Loop melalui semua folder di dataset
for subdir in os.listdir(dataset_dir):
    subdir_path = os.path.join(dataset_dir, subdir)

    # Periksa apakah subdir adalah folder
    if os.path.isdir(subdir_path):
        # Buat folder output untuk setiap subdir
        subdir_output_dir = os.path.join(output_dir, subdir)
        os.makedirs(subdir_output_dir, exist_ok=True)

        # Loop untuk resize setiap gambar di folder
        for filename in os.listdir(subdir_path):
            if filename.endswith('.jpeg') or filename.endswith('.jpg') or filename.endswith('.png'):
                img_path = os.path.join(subdir_path, filename)
                img = Image.open(img_path)

                # Resize gambar
                img_resized = img.resize(new_size)

                # Simpan gambar yang sudah di-resize
                output_path = os.path.join(subdir_output_dir, filename)
                img_resized.save(output_path)

print("Resize selesai. Semua folder diproses, hasilnya ada di folder:", output_dir)
```

```python
from tensorflow.keras import layers

batch_size = 32
img_size = (224,224)
dataset_path = './resized_dataset'

train_dataset = tf.keras.utils.image_dataset_from_directory(
    dataset_path,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=img_size,
    batch_size=batch_size
)

val_dataset = tf.keras.utils.image_dataset_from_directory(
    dataset_path,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=img_size,
    batch_size=batch_size
)

# Augmentation
data_augmentation = tf.keras.Sequential([
    layers.Rescaling(1./255),
    layers.RandomFlip("horizontal_and_vertical"),
    layers.RandomRotation(0.2),
    layers.RandomZoom(0.2),
    layers.RandomTranslation(0.2, 0.2),
])

# Terapkan augmentation
augmented_train_dataset = train_dataset.map(
    lambda x, y: (data_augmentation(x, training=True), y)
)

# # Prefetching
# train_dataset = augmented_train_dataset.prefetch(buffer_size=tf.data.AUTOTUNE)
# val_dataset = val_dataset.prefetch(buffer_size=tf.data.AUTOTUNE)
```

# **NLP Food Menus Recommendation**

```python
import kagglehub
import pandas as pd

path = kagglehub.dataset_download("albertnathaniel12/food-recipes-dataset")
print("Path to dataset files:", path)

csv_file = f"{path}/Indonesian_Food_Recipes.csv"

df = pd.read_csv(csv_file)
df.head()
```

```python
import pandas as pd
import tensorflow as tf
import numpy as np

dataset_ingredients = df["Ingredients Cleaned"].astype(str).tolist()

# TextVectorization
max_vocab_size = 10000
sequence_length = 50

vectorizer = tf.keras.layers.TextVectorization(
    max_tokens=max_vocab_size,
    output_mode="int",
    output_sequence_length=sequence_length,
)


vectorizer.adapt(dataset_ingredients)
encoded_dataset = vectorizer(dataset_ingredients).numpy()

# Similarity Function
def cosine_similarity(a, b):
    a = tf.constant(a, dtype=tf.float32)
    b = tf.constant(b, dtype=tf.float32)
    dot_product = tf.reduce_sum(a * b, axis=1)
    norm_a = tf.sqrt(tf.reduce_sum(a**2, axis=1))
    norm_b = tf.sqrt(tf.reduce_sum(b**2, axis=1))
    similarity = dot_product / (norm_a * norm_b)
    return similarity

# Set of ingredients (user-provided)
input_ingredients = "ayam, cabai merah, cabai hijau"

# Encode the input ingredients
encoded_input = vectorizer([input_ingredients]).numpy()
# Compute cosine similarity
similarity_scores = cosine_similarity(encoded_dataset, encoded_input).numpy()
# Top 3 most similar menus
top_3_indices = np.argsort(similarity_scores)[-3:][::-1]  # Sort in descending order
top_3_menus = df.iloc[top_3_indices]

# Results
print("Input Ingredients:", input_ingredients)
print("\nTop 3 Similar Menus:")
for i, (index, row) in enumerate(top_3_menus.iterrows(), start=1):
    print(f"{i}. Menu: {row['Title']}")
    print(f"   Ingredients: {row['Ingredients Cleaned']}")
    print(f"   Similarity Score: {similarity_scores[index]:.4f}")

# ini masih questionnable outputnya, kayanya harus coba metrics jarak selain cosine similarity
```
